```python
import pandas as pd

# 1. Đọc file dữ liệu của bạn (thay đổi tên file và đường dẫn cho đúng)
# Nếu là file Excel: pd.read_excel('ten_file.xlsx')
df = pd.read_csv('vismishds_phase1_final.csv') 

# 2. Lọc ra các giá trị xuất hiện từ 2 lần trở lên trong cột 'content'
# duplicated(keep=False) sẽ đánh dấu tất cả các dòng bị trùng
trung_lap = df[df.duplicated(subset=['content'], keep=False)]

# 3. Gom nhóm, đếm số lần xuất hiện và sắp xếp giảm dần để tiện theo dõi
thong_ke_trung = (
    trung_lap.groupby('content')
    .size()
    .reset_index(name='so_lan_trung')
    .sort_values(by='so_lan_trung', ascending=False)
)

# 4. Hiển thị kết quả
print(f"Tổng số tin nhắn bị trùng lặp: {len(thong_ke_trung)}")
print("-" * 50)
print(thong_ke_trung.head(20)) # Xem top 20 giá trị trùng nhiều nhất
```

```python
# 1. Đọc file dữ liệu gốc
df = pd.read_csv('vismishds_phase1_final.csv') 
print(f"Số dòng ban đầu: {len(df)}")

# 2. Xóa các dòng có 'content' bị trùng, chỉ giữ lại dòng đầu tiên
# Tham số keep='first' có nghĩa là giữ lại dòng đầu và xóa các dòng trùng phía sau
df.drop_duplicates(subset=['content'], keep='first', inplace=True)

print(f"Số dòng sau khi xóa trùng: {len(df)}")
print(f"Đã xóa được: {len(df) - len(df)} dòng trùng lặp.")

# 3. Lưu kết quả ra file mới để sử dụng
df.to_csv('vismishds_phase1_final_clean.csv', index=False)
print("Đã lưu dữ liệu sạch vào file 'vismishds_phase1_final_clean.csv'")
```

```python
# 1. Đọc file dữ liệu (thay đổi tên file cho đúng với dữ liệu của bạn)
df = pd.read_csv('vismishds_phase1_final_clean.csv') 

# 2. Lọc dữ liệu chỉ lấy các dòng có label bằng 1
df_filtered = df[df['label'] == 1]

# 3. Thống kê số lượng từng giá trị trong cột 'obfuscation_level'
thong_ke_so_luong = df_filtered['obfuscation_level'].value_counts()

# 4. Thống kê thêm tỷ lệ phần trăm (%) để bạn dễ đánh giá (tùy chọn)
thong_ke_phan_tram = df_filtered['obfuscation_level'].value_counts(normalize=True) * 100

# 5. In kết quả ra màn hình
print("=== THỐNG KÊ SỐ LƯỢNG (Label = 1) ===")
print(thong_ke_so_luong)

print("\n=== THỐNG KÊ TỶ LỆ PHẦN TRĂM (%) ===")
# Giới hạn hiển thị 2 chữ số thập phân cho đẹp
print(thong_ke_phan_tram.round(2).astype(str) + '%')
```

```python
#Thay thế toàn bộ giá trị 'NONE' thành 'LEVEL 0 – Không obfuscation (formal)'
df_filtered['obfuscation_level'] = df_filtered['obfuscation_level'].replace({
    'NONE': 'LEVEL 0 – Không obfuscation (formal)'
})

print("=== THỐNG KÊ SAU KHI GỘP GIÁ TRỊ (Label = 1) ===")
print(df_filtered['obfuscation_level'].value_counts())

df.to_csv('vismishds_phase1_final_clean.csv', index=False)
```

```python
nhom_012 = [
    'LEVEL 0 – Không obfuscation (formal)',
    'LEVEL 1 – Leet nhẹ (thay 1-2 ký tự)',
    'LEVEL 2 – Leet nặng + tên riêng'
]

nhom_345 = [
    'LEVEL 3 – Dot/dash insertion',
    'LEVEL 4 – Mixed special chars',
    'LEVEL 5 – Extreme noise'
]

df_012 = df_filtered[df_filtered['obfuscation_level'].isin(nhom_012)]
df_345 = df_filtered[df_filtered['obfuscation_level'].isin(nhom_345)]
print(f"Tổng số dòng của dữ liệu gốc: {len(df_filtered)}")
print(f"Số dòng của ds012 (Level 0, 1, 2): {len(df_012)}")
print(f"Số dòng của ds345 (Level 3, 4, 5): {len(df_345)}")

#Xuất ra 2 file CSV riêng biệt
df_012.to_csv('ds012.csv', index=False)
df_345.to_csv('ds345.csv', index=False)
```
