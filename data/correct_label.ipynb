```python
import pandas as pd

input_path = r"c:\KLTN\KLTN\data\ground_truth\dataset_label_1_with_categories.csv"
output_path = r"c:\KLTN\KLTN\data\ground_truth\khac_smishing_type.csv"

df = pd.read_csv(input_path)

khac_df = df[df["smishing_type"].astype(str).str.strip() == "Khác"]

khac_df = khac_df[["content","label","has_url","has_phone_number","sender_type"]]
# Lưu file mới
khac_df.to_csv(output_path, index=False, encoding="utf-8-sig")

print(f"Da luu {len(khac_df)} dong vao: {output_path}")
```

```python
import pandas as pd

path_main = r"c:\KLTN\KLTN\data\ground_truth\dataset_label_1_with_categories.csv"
path_khac = r"c:\KLTN\KLTN\data\ground_truth\khac_smishing_type.csv"

df_main = pd.read_csv(path_main)
df_khac = pd.read_csv(path_khac)

# Chuẩn hóa content để so khớp ổn định hơn
norm_main = df_main["content"].astype(str).str.strip()
norm_khac_set = set(df_khac["content"].astype(str).str.strip())

is_khac = df_main["smishing_type"].astype(str).str.strip().eq("Khác")
keep_khac = is_khac & norm_main.isin(norm_khac_set)
keep_non_khac = ~is_khac

df_filtered = df_main[keep_non_khac | keep_khac].copy()

print("Tong truoc loc:", len(df_main))
print("So dong 'Khác' truoc loc:", is_khac.sum())
print("So dong 'Khác' giu lai:", keep_khac.sum())
print("Tong sau loc:", len(df_filtered))

# Ghi đè file gốc:
df_filtered.to_csv(path_main, index=False, encoding="utf-8-sig")

# Nếu muốn an toàn hơn, lưu file mới thay vì ghi đè:
# out_path = r"c:\KLTN\KLTN\data\ground_truth\dataset_label_1_with_categories_filtered.csv"
# df_filtered.to_csv(out_path, index=False, encoding="utf-8-sig")
```

```python
import pandas as pd

path_src = r"c:\KLTN\KLTN\data\ground_truth\dataset_label_1_with_categories.csv"
path_dst = r"c:\KLTN\KLTN\data\ground_truth\dataset_label_1.csv"

df = pd.read_csv(path_src)

# Bỏ smishing_type nếu có
df_out = df.drop(columns=["smishing_type"], errors="ignore")

# (tuỳ chọn) đảm bảo đúng thứ tự cột chuẩn của dataset_label_1
expected_cols = ["content", "label", "has_url", "has_phone_number", "sender_type"]
df_out = df_out[expected_cols]

# Ghi đè dataset_label_1.csv
df_out.to_csv(path_dst, index=False, encoding="utf-8-sig")

print("Da cap nhat dataset_label_1.csv")
print("So dong:", len(df_out))
print("Cac cot:", list(df_out.columns))
```
