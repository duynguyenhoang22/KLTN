```python
import pandas as pd
```

```python
df = pd.read_csv('phase2_full_normalization_final.csv')
```

```python
# Thay thế tất cả cụm "BHXH VN" trong cột normalized_text thành "Bảo hiểm Xã hội Việt Nam"
df['normalized_text'] = df['normalized_text'].str.replace('BHXH', 'Bảo hiểm Xã hội', regex=False)
```

```python
df['normalized_text'] = df['normalized_text'].str.replace('MTTQ', 'Mặt trận Tổ quốc', regex=False)
```

```python
df['normalized_text'] = df['normalized_text'].str.replace(' VN ', ' Việt Nam ', regex=False)
```

```python
df['normalized_text'] = df['normalized_text'].str.replace(' CNTT ', ' Công nghệ Thông tin ', regex=False)
```

```python
df['normalized_text'] = df['normalized_text'].str.replace('CSKH', 'Chăm sóc khách hàng', regex=False)
```

```python
df['normalized_text'] = df['normalized_text'].str.replace('BHYT', 'Bảo hiểm Y tế', regex=False)
```

```python
df['normalized_text'] = df['normalized_text'].str.replace(' LH ', ' liên hệ ', regex=False)
```

```python
df['normalized_text'] = df['normalized_text'].str.replace('TCT', 'Tổng cục Thuế', regex=False)
```

```python
df['normalized_text'] = df['normalized_text'].str.replace('TP.HCM', 'Thành phố Hồ Chí Minh', regex=False)
```

```python
df['normalized_text'] = df['normalized_text'].str.replace('[VTP]', '[Viettel Post]', regex=False)
```

## GHTK - Giao Hàng Tiết Kiệm

```python
df['normalized_text'] = df['normalized_text'].str.replace('[GHTK]', '[Giao Hàng Tiết Kiệm]', regex=False)
```

```python
df['normalized_text'] = df['normalized_text'].str.replace('GHTK:', 'Giao Hàng Tiết Kiệm:', regex=False)
```

```python
df['normalized_text'] = df['normalized_text'].str.replace('GHTK-', 'Giao Hàng Tiết Kiệm-', regex=False)
```

## GHN - Giao Hàng Nhanh

```python
df['normalized_text'] = df['normalized_text'].str.replace('[GHN]', '[Giao Hàng Nhanh]', regex=False)
```

```python
df['normalized_text'] = df['normalized_text'].str.replace('GHN:', 'Giao Hàng Nhanh:', regex=False)
```

```python
# Lưu lại file
df.to_csv('phase2_full_normalization_final.csv', index=False)
```
